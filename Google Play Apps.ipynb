# Google Play Store Apps
Web scraped data of 10k Play Store apps for analysing the Android market
A dataset provided by [kaggle](https://www.kaggle.com/lava18/google-play-store-apps#googleplaystore.csv)

```python
df['Price']=df['Price'].apply(lambda x: x.replace('$',''))
```

```python
df=df[df.Price!='Everyone' ]
```

```python
df['Price']=df['Price'].astype(float)
```

```python
df['Category']=df['Category'].apply(lambda x: x.split('_')[0])
```

## Grouping Category and Price

```python
amount_spentbyUserOn_Each_Category=df.groupby('Category')['Price'].sum()
amount_spentbyUserOn_Each_Category
```

## Visualization of Categories of Applications with Price

```python
plt.figure(figsize=(40,20))
sns.set(rc={'figure.figsize':(35.7,28.27)})
sns.set(font_scale = 2)
#sns.set_xticklabels( x=money.index, rotation=60 )
plt.draw()
chart=sns.barplot(x=amount_spentbyUserOn_Each_Category.index,y=amount_spentbyUserOn_Each_Category.values)

chart.set_xticklabels(labels=chart.get_xticklabels(), rotation=90, horizontalalignment='right')

```
